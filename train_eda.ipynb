```python
# Correlation all column

# string 형식 column 제거
corr_data = data.drop(columns = ["자치구", "행정동"])

# 한글 폰트 및 -폰트 깨지는 현상 제거
mpl.rcParams["font.family"] = "NanumGothic"
sns.set(font = "NanumGothic", rc = {"axes.unicode_minus": False}, style = "ticks")

# correlation 계산 및 plotting
corr = corr_data.corr()
fig = plt.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(corr, cmap = 'coolwarm', vmin=-1, vmax=1)
fig.colorbar(cax)
ticks = np.arange(0,len(corr_data.columns),1)
ax.set_xticks(ticks)
plt.xticks(rotation=90)
ax.set_yticks(ticks)
ax.set_xticklabels(corr_data.columns)
ax.set_yticklabels(corr_data.columns)
plt.show()
```

연령대로 묶어서 분석하기

```python
groupedAge = data.drop(columns = ["자치구", "행정동"]).groupby(["연령대"]).sum()
```

```python
groupedAge
```

퍼센트 비율 버전

```python
# 50대가 다른 연령대 보다 인원 수가 2배로 많음
# 연령대별 비율을 구하기 위해 각 column을 연령대 총 인구수로 나눔

# 연령대별 인구수
total = groupedAge["총인구수"]
t20 = total.iloc[0]
t30 = total.iloc[1]
t40 = total.iloc[2]
t50 = total.iloc[3]

# 평균값은 총 인구수에 영향을 받지 않기 때문에 비율화 진행하지 않음
# 해당 column 그대로 가져가기 위해 list로 저장
avg = list(groupedAge["휴일 총 이동 횟수 평균"])

# 평균값 column 남겨두고 비율 테이블 생성
ratio_data_raw = groupedAge.drop(columns = "휴일 총 이동 횟수 평균")

# 위에서 구한 연령대 총 인구수로 scaling 진행
ratio_data_20 = ratio_data_raw.iloc[0].div(t20)
ratio_data_30 = ratio_data_raw.iloc[1].div(t30)
ratio_data_40 = ratio_data_raw.iloc[2].div(t40)
ratio_data_50 = ratio_data_raw.iloc[3].div(t50)
ratio_data_list = [ratio_data_20, ratio_data_30, ratio_data_40, ratio_data_50]

# dataframe으로 생성
ratio_data = pd.DataFrame(data = ratio_data_list, index = None)

# 값이 너무 작으니 percent로 
per_data = ratio_data.mul(100)

# 연령대 살려오기 위해 index를 reset하고 이름을 연령대로 변경
per_data = per_data.reset_index()
per_data.rename(columns = {"index" : "연령대"}, inplace = True)

# 평균 값 df에 concat
per_data["휴일 총 이동 횟수 평균"] = avg
per_data
```

```python
sns.barplot(x = "연령대", y = "평일 총 이동 횟수", data = per_data)
```

```python
sns.barplot(x = "연령대", y = "평일 총 이동 거리 합계", data = per_data)
```
