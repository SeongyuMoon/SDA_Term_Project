퍼센트 비율 버전

```python
# 50대가 다른 연령대 보다 인원 수가 2배로 많음
# 연령대별 비율을 구하기 위해 각 column을 연령대 총 인구수로 나눔

# 연령대별 인구수
total = groupedAge["총인구수"]
t20 = total.iloc[0]
t30 = total.iloc[1]
t40 = total.iloc[2]
t50 = total.iloc[3]
t60 = total.iloc[4]

# 평균값은 총 인구수에 영향을 받지 않기 때문에 비율화 진행하지 않음
# 해당 column 그대로 가져가기 위해 list로 저장
avg = list(groupedAge["휴일 총 이동 횟수 평균"])

# 평균값 column 남겨두고 비율 테이블 생성
ratio_data_raw = groupedAge.drop(columns = "휴일 총 이동 횟수 평균")

# 위에서 구한 연령대 총 인구수로 scaling 진행
ratio_data_20 = ratio_data_raw.iloc[0].div(t20)
ratio_data_30 = ratio_data_raw.iloc[1].div(t30)
ratio_data_40 = ratio_data_raw.iloc[2].div(t40)
ratio_data_50 = ratio_data_raw.iloc[3].div(t50)
ratio_data_60 = ratio_data_raw.iloc[4].div(t60)
ratio_data_list = [ratio_data_20, ratio_data_30, ratio_data_40, ratio_data_50, ratio_data_60]

# dataframe으로 생성
ratio_data = pd.DataFrame(data = ratio_data_list, index = None)

# 값이 너무 작으니 percent로 
per_data = ratio_data.mul(100)

# 연령대 살려오기 위해 index를 reset하고 이름을 연령대로 변경
per_data = per_data.reset_index()
per_data.rename(columns = {"index" : "연령대"}, inplace = True)

# 평균 값 df에 concat
per_data["휴일 총 이동 횟수 평균"] = avg
per_data
```

```python
from scipy import stats
import statsmodels.api as sm
from scipy.stats import shapiro
from scipy.stats import kstest
from scipy.stats import bartlett
from scipy.stats import levene
from scipy.stats import fligner
from scipy.stats import chi2
from scipy.stats import chisquare
from scipy.stats import chi2_contingency
from scipy.stats import f_oneway
from scipy.stats import f
from statsmodels.stats.multitest import multipletests
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import statsmodels.api as sm
```

<h4>Hypothesis 1</h4>
<h5>H0: 각 연령대 별 평일 총 이동 횟수는 차이가 없다.</h5>
<h5>HA: 각 연령대 별 평일 총 이동 횟수는 차이가 있다.</h5>

```python
weekday_move_20 = data[data["연령대"] == 20]["평일 총 이동 횟수"]
weekday_move_30 = data[data["연령대"] == 30]["평일 총 이동 횟수"]
weekday_move_40 = data[data["연령대"] == 40]["평일 총 이동 횟수"]
weekday_move_50 = data[data["연령대"] == 50]["평일 총 이동 횟수"]
weekday_move_60 = data[data["연령대"] == 60]["평일 총 이동 횟수"]
```

```python
# 20대 평일 총 이동 횟수 정규성 검정
fig = sm.qqplot(weekday_move_20, line = "45")
```

```python
shapiro(weekday_move_20)
```

```python
# 30대 평일 총 이동 횟수 정규성 검정
fig = sm.qqplot(weekday_move_30, line = "45")
```

```python
shapiro(weekday_move_30)
```

```python
# 40대 평일 총 이동 횟수 정규성 검정
fig = sm.qqplot(weekday_move_40, line = "45")
```

```python
shapiro(weekday_move_40)
```

```python
# 50대 평일 총 이동 횟수 정규성 검정
fig = sm.qqplot(weekday_move_50, line = "45")
```

```python
shapiro(weekday_move_50)
```

```python
# 60대 이상 평일 총 이동 횟수 정규성 검정
fig = sm.qqplot(weekday_move_60, line = "45")
```

```python
shapiro(weekday_move_60)
```

모든 연령대 전부 정규성 띄지 않음

```python
# 동분산 검정
levene(weekday_move_20, weekday_move_30, weekday_move_40, weekday_move_50, weekday_move_60)
```

```python
bartlett(weekday_move_20, weekday_move_30, weekday_move_40, weekday_move_50, weekday_move_60)
```

```python
fligner(weekday_move_20, weekday_move_30, weekday_move_40, weekday_move_50, weekday_move_60)
```

모두 정규성을 띄지 않아 levene이 가장 강력한 결과를 보일 것.
<h5>근데 다 p < 0.05 --> 동분산 X</h5>

```python
# kruskal test(ANOVA 진행하기는 정규성, 동분산성 만족하지 않음)
_, p = stats.kruskal(weekday_move_20, weekday_move_30, weekday_move_40, weekday_move_50, weekday_move_60)
print(f"kruskal test p-value: {p}")
```

p < 0.05 --> 최소한 1개 이상의 집단에서 차이

```python
# 각 연령대 별 test
tukey = pairwise_tukeyhsd(endog = data["평일 총 이동 횟수"], groups = data["연령대"], alpha = 0.05)
print(tukey)
```

30대와 40대의 평일 총 이동 횟수의 평균의 차이 제외하고는 모든 집단에서 차이
